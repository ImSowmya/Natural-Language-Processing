```python
words = ["running", "runs", "stemming", "easily", "fairly", "cars", "children", "wolves", "happier", "happiness", "studies"]
```

```python
import nltk
# nltk.download('punct')
```

```python
from nltk.stem import PorterStemmer
stemming = PorterStemmer()
ps_words = [[word,stemming.stem(word)] for word in words]
ps_words
```

```python
from nltk.stem import LancasterStemmer
stemming = LancasterStemmer()
ls_words = [[word,stemming.stem(word)] for word in words]
ls_words
```

```python
from nltk.stem import SnowballStemmer
stemming = SnowballStemmer('english')
sbs_words = [[word,stemming.stem(word)] for word in words]
sbs_words
```

```python
from nltk.stem import RegexpStemmer
stemming = RegexpStemmer('ing$|s$|e$|able$', min=4)
regexp_words = [[word,stemming.stem(word)] for word in words]
regexp_words
```

```python
import pandas as pd
stemming_df = pd.DataFrame()

stemming_df['Original_words'] = words
stemming_df['RegExpStemmer'] = [RegexpStemmer('ing$|s$|e$|able$', min=4).stem(word) for word in words]
stemming_df['PorterStemmer'] = [PorterStemmer().stem(word) for word in words]
stemming_df['SnowballStemmer'] = [SnowballStemmer('english').stem(word) for word in words]
stemming_df['LancasterStemmer'] = [LancasterStemmer().stem(word) for word in words]

stemming_df
```

```python
from nltk.stem import WordNetLemmatizer
#nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
lemma_words = [lemmatizer.lemmatize(word,'a') for word in words]
```

```python
stemming_df["Lemmatizer"] = lemma_words
stemming_df
```
